```python
from keras.layers.core import Dense, SpatialDropout1D
from keras.layers.convolutional import Conv1D
from tensorflow.keras.layers import Embedding
from keras.layers.pooling import GlobalMaxPooling1D
from keras.models import Sequential
from tensorflow.keras.preprocessing.sequence import pad_sequences
from keras.utils import np_utils
from sklearn.model_selection import train_test_split
import collections
import nltk
import numpy as np
import codecs
import string as st
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
def remove_stopwords(text):
    stopword=nltk.corpus.stopwords.words('english')
    stopword.remove('not')
    w=[word for word in nltk.word_tokenize(text) if word not in stopword]
    return ' '.join(w)
```

```python
def remove_punctuation(text):
    w=[word for word in text if word not in st.punctuation]
    return ' '.join(w)
```

```python
INPUT_FILE = "/content/train_2k.txt"
```

##**Dataset Preparation and Cleaning**

```python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer()
counter = collections.Counter()
maxlen = 0
xs, ys = [], []



fin = codecs.open(INPUT_FILE, "r", encoding='utf-8')
for line in fin:
  
    label, sent = line.strip().split("\t") #Stripping the dataset based on tab. That is stripping label from sentence
    print("Label: ", label)
    print("Sentence: ",sent)
    
    words = nltk.word_tokenize(sent)
    print("Afert Tokenizing: ",words)

    wordsExcludingPunctuationMarks=remove_punctuation(words)
    print("Truncating punctuation:", wordsExcludingPunctuationMarks)

    wordsExcludingStopWords=remove_stopwords(wordsExcludingPunctuationMarks)
    print("Truncating StopWords:", wordsExcludingStopWords)

    if len(wordsExcludingStopWords) > maxlen: #For calculating the maximum number of words in a sentence
        maxlen = len(wordsExcludingStopWords) 
    for wordExcludingStopWords in wordsExcludingStopWords:
        counter[wordExcludingStopWords] += 1 #Putting the frequency of each  word in a dictionary
    print("***************************************************************************************")

    ys.append(int(label))
    xs.append(''.join(wordsExcludingStopWords))


fin.close()
    
```

```python
xs
```

##**TF-IDF**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(xs, ys, test_size=0.20, random_state=0)
Xtrain
```

```python
#tf idf
tf_idf = TfidfVectorizer()
#applying tf idf to training data
Xtrain_tf = tf_idf.fit_transform(Xtrain)
#applying tf idf to training data
Xtrain_tf = tf_idf.transform(Xtrain)
print("n_samples: %d, n_features: %d" % Xtrain_tf.shape)
```

```python
#transforming test data into tf-idf matrix
Xtest_tf = tf_idf.transform(Xtest)
print("n_samples: %d, n_features: %d" % Xtest_tf.shape)
```

##**Naive Bayes**

```python
#naive bayes classifier
from sklearn.naive_bayes import MultinomialNB
naive_bayes_classifier = MultinomialNB()
naive_bayes_classifier.fit(Xtrain_tf, Ytrain)
#predicted y
y_pred = naive_bayes_classifier.predict(Xtest_tf)
```

```python

from sklearn import metrics
print(metrics.classification_report(Ytest, y_pred, target_names=['Positive', 'Negative']))
```

```python
print(metrics.confusion_matrix(Ytest, y_pred))
```

##**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
LRClassification=LogisticRegression()
LRClassification.fit(Xtrain_tf, Ytrain)
y_pred=LRClassification.predict(Xtest_tf)
```

```python

from sklearn import metrics
print(metrics.classification_report(Ytest, y_pred, target_names=['Positive', 'Negative']))
```

```python
print(metrics.confusion_matrix(Ytest, y_pred))
```

```python
fprLogistic, tprLogistic, thresholdsLogistic = metrics.roc_curve(Ytest,y_pred)
```

##**Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier
DTClassification=DecisionTreeClassifier(random_state=0)
DTClassification.fit(Xtrain_tf, Ytrain)
y_pred=DTClassification.predict(Xtest_tf)
```

```python
from sklearn import metrics
print(metrics.classification_report(Ytest, y_pred, target_names=['Positive', 'Negative']))
```

```python
print(metrics.confusion_matrix(Ytest, y_pred))
```

```python
fprDecision, tprDecision, thresholdsDecision = metrics.roc_curve(Ytest,y_pred)
```

##**SVM**
